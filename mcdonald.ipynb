```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from sklearn.utils import resample
import matplotlib.pyplot as plt

np.random.seed(1234)

#k-means clustering and return the cluster labels
def kmeans_clustering(data, n_clusters, n_init=10):
    kmeans = KMeans(n_clusters=n_clusters, n_init=n_init, random_state=1234)
    kmeans.fit(data)
    return kmeans.labels_

def compute_global_stability(data, min_clusters=2, max_clusters=8, n_bootstrap=100, n_init=10):
    stability_scores = {n_clusters: [] for n_clusters in range(min_clusters, max_clusters + 1)}

    for n_clusters in range(min_clusters, max_clusters + 1):
        for _ in range(n_bootstrap):
            bootstrap_sample = resample(data, random_state=1234)
            #k-means clustering on the bootstrap sample
            bootstrap_labels = kmeans_clustering(bootstrap_sample, n_clusters, n_init)
            #Compare to the clustering on the original data
            original_labels = kmeans_clustering(data, n_clusters, n_init)
            score = adjusted_rand_score(original_labels, bootstrap_labels)
            stability_scores[n_clusters].append(score)
    return stability_scores

#Compute global stability
stability_scores = compute_global_stability(df2, n_bootstrap=100, n_init=10)

#results
plt.figure(figsize=(10, 6))
for n_clusters, scores in stability_scores.items():
    plt.boxplot(scores, positions=[n_clusters], widths=0.5, labels=[n_clusters])
plt.xlabel('Number of segments')
plt.ylabel('Adjusted Rand Index')
plt.title('Global Stability Boxplot')
#plt.ylim(-0.01,0.025)
plt.show()
```

```python
from sklearn.metrics import pairwise_distances
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances
MD_x=df2
#Function to compute pairwise distances within each segment
def compute_distances_within_segments(data, labels, n_segments):
    distances_within_segments = {i: [] for i in range(n_segments)}
    #Calculate the distance matrix
    distances = pairwise_distances(data, data, metric='euclidean')
    #Collect distances within each segment
    for i in range(n_segments):
        segment_points = data[labels == i]
        if len(segment_points) > 1:  # Ensure there are at least two points in the segment
            segment_distances = distances[labels == i][:, labels == i]
            upper_tri_indices = np.triu_indices_from(segment_distances, k=1)
            distances_within_segments[i] = segment_distances[upper_tri_indices]
    return distances_within_segments
#Fit k-means model for 4 segments
kmeans_4 = KMeans(n_clusters=4, random_state=1234)
kmeans_4.fit(MD_x)
labels_4_segments = kmeans_4.labels_
# Compute pairwise distances within each segment
distances_within_segments = compute_distances_within_segments(MD_x, labels_4_segments, 4)

fig, axes = plt.subplots(2, 2, figsize=(12, 10), sharex=True, sharey=True)

for i, (segment, distances) in enumerate(distances_within_segments.items()):
    ax = axes[i // 2, i % 2]
    ax.hist(distances, bins=30, range=(0, 1), edgecolor='black')
    ax.set_title(f'Segment {i + 1}')
    ax.set_xlabel('Distance')
    ax.set_ylabel('Frequency')
    ax.set_xlim(0, 1)
    ax.grid(True)

plt.tight_layout()
plt.show()
```

Profiling segment

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.spatial.distance import pdist
MD_x=df2
#hierarchical clustering
distance_matrix = pdist(MD_x.T, metric='euclidean')
linkage_matrix = linkage(distance_matrix, method='ward')

plt.figure(figsize=(10, 7))
dendrogram(linkage_matrix, labels=np.arange(MD_x.shape[1]), leaf_rotation=90)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Feature')
plt.ylabel('Distance')
plt.show()
```

```python
pca = PCA(n_components=2)
MD_pca = pca.fit_transform(MD_x)

# Plot PCA projection
plt.figure(figsize=(10, 8))
scatter = plt.scatter(MD_pca[:, 0], MD_pca[:, 1], cmap='viridis', alpha=0.6)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Projection of Segments')
plt.colorbar(scatter, label='Segment')
plt.grid(True)
plt.show()
# Plot PCA projection with axes
plt.figure(figsize=(10, 8))
scatter = plt.scatter(MD_pca[:, 0], MD_pca[:, 1], cmap='viridis', alpha=0.6)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Projection of Segments with Axes')

# Add PCA vectors (axes)
for i in range(pca.components_.shape[0]):
    plt.arrow(0, 0, pca.components_[i, 0], pca.components_[i, 1], head_width=0.1, head_length=0.1, fc='r', ec='r')

plt.colorbar(scatter, label='Segment')
plt.grid(True)
plt.show()
```

Describing segments

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Generate example data
np.random.seed(1234)
MD_x = df2
kmeans_4 = KMeans(n_clusters=4, random_state=1234)
kmeans_4.fit(MD_x)
cluster_labels = kmeans_4.labels_

# Create DataFrame with matching lengths
mcdonalds = pd.DataFrame({
    'Segment': cluster_labels,
    'Like': np.random.choice(['Yes', 'No'], size=MD_x.shape[0]),
    'Gender': np.random.choice(['Male', 'Female'], size=MD_x.shape[0]),
    'Age': np.random.randint(18, 65, size=MD_x.shape[0]),
    'VisitFrequency': np.random.randint(1, 10, size=MD_x.shape[0])
})

# Verify lengths
print(f"Segment length: {len(mcdonalds['Segment'])}")
print(f"Like length: {len(mcdonalds['Like'])}")
print(f"Gender length: {len(mcdonalds['Gender'])}")

# Distribution of 'Like' across segments
plt.figure(figsize=(10, 6))
sns.countplot(data=mcdonalds, x='Segment', hue='Like', palette='coolwarm')
plt.title('Distribution of Likes Across Segments')
plt.xlabel('Segment Number')
plt.ylabel('Count')
plt.legend(title='Like')
plt.show()

# Distribution of 'Gender' across segments
plt.figure(figsize=(10, 6))
sns.countplot(data=mcdonalds, x='Segment', hue='Gender', palette='coolwarm')
plt.title('Distribution of Gender Across Segments')
plt.xlabel('Segment Number')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()

# Prepare data for decision tree
X = mcdonalds[['Like', 'Age', 'VisitFrequency', 'Gender']]
X = pd.get_dummies(X, drop_first=True)  # Convert categorical variables to dummy/indicator variables
y = mcdonalds['Segment']

# Check lengths
print(f"Feature matrix rows: {X.shape[0]}, Target length: {len(y)}")

# Train decision tree classifier
clf = DecisionTreeClassifier(random_state=1234)
clf.fit(X, y)

# Plot decision tree
plt.figure(figsize=(20, 10))
plot_tree(clf, feature_names=X.columns, class_names=[f'Segment {i + 1}' for i in range(4)], filled=True)
plt.title('Decision Tree for Segment Description')
plt.show()

```

Selecting target segment

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate example data
np.random.seed(1234)
MD_x = df2  # Example data
kmeans_4 = KMeans(n_clusters=4, random_state=1234)
kmeans_4.fit(MD_x)
cluster_labels = kmeans_4.labels_

mcdonalds = pd.DataFrame({
    'Segment': cluster_labels,
    'Like': np.random.choice([0, 1], size=MD_x.shape[0]),  # 0 or 1 for Like
    'Gender': np.random.choice(['Male', 'Female'], size=MD_x.shape[0]),
    'VisitFrequency': np.random.randint(1, 10, size=MD_x.shape[0])
})

# Calculate mean values for each segment
visit = mcdonalds.groupby('Segment')['VisitFrequency'].mean()
like = mcdonalds.groupby('Segment')['Like'].mean()
female = mcdonalds.groupby('Segment')['Gender'].apply(lambda x: (x == 'Female').mean())

print('Visit Frequency by Segment:', visit)
print('Like by Segment:', like)
print('Female Proportion by Segment:', female)
```

```python
# Plot
plt.figure(figsize=(10, 8))
plt.scatter(visit, like, s=50 * female, alpha=0.7, c='blue', edgecolor='black')

# Annotate points with segment numbers
for segment in visit.index:
    plt.text(visit[segment], like[segment], str(segment + 1), fontsize=12, ha='right')

plt.xlabel('Average Visit Frequency')
plt.ylabel('Average Like')
plt.xlim(visit.min() - 1, visit.max() + 1)  # Adjust limits as needed
plt.ylim(like.min() - 0.5, like.max() + 0.5)  # Adjust limits as needed
plt.title('Segment Analysis')
plt.grid(True)
plt.show()
```
